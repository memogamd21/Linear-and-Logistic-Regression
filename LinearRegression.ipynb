```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import pandas as pd
```

```python
def create_basis(x, degree=1):
    x = np.asarray(x) #a 1d array
    X = [x**i for i in range(0,degree+1)] #include 0 as we want a column of 1s
    X = np.asarray(X).T
    return X
```
